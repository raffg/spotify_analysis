```python
df = pd.concat([df1, df2, df3, df4, df5, df6])
df.drop_duplicates(inplace=True)
print(len(df))
```

```python
countries = [
    "global", "us", "gb", "ad", "ar", "at", "au", "be", "bg", "bo", "br", "ca",
    "ch", "cl", "co", "cr", "cy", "cz", "de", "dk", "do", "ec", "ee", "es",
    "fi", "fr", "gr", "gt", "hk", "hn", "hu", "id", "ie", "il", "in", "is",
    "it", "jp", "lt", "lu", "lv", "mc", "mt", "mx", "my", "ni", "nl", "no",
    "nz", "pa", "pe", "ph", "pl", "pt", "py", "ro", "se", "sg", "sk", "sv",
    "th", "tr", "tw", "uy", "vn", "za"
]
```

```python
from datetime import date, timedelta
initial_date = date(2017, 1, 1)
next_date = timedelta(days=1)
temp_date = initial_date
dates = []
while temp_date <= date.today():
    dates.append(temp_date)
    temp_date += next_date
```

```python
temp = df1.groupby(['Date', 'Country'])['Position'].count()
missing1 = []
for country in countries:
    for date in dates:
        try:
            temp.loc[str(date), country]
        except KeyError:
            missing1.append((date, country))
print(len(missing1))
country_count1 = []
for country in set([item[1] for item in missing1]):
    country_count1.append((country, len([item for item in missing1 if item[1] == country])))
sorted(country_count1, key=lambda tup: tup[1])
```
